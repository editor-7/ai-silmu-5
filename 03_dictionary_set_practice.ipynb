## 4. 집합으로 중복 제거하기

```python
# TODO 3: interests에 중복 값을 하나 넣고, set으로 중복 제거를 확인하세요.
interests = ["자동화", "데이터", "자동화", "리포트"]
unique_interests = set(interests)    # interests 변수를 set을 활용해서 새로운 unique_interests 만들기
print("원래 리스트:", interests)
print("중복 제거:", unique_interests)
```
